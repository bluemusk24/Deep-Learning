# 8.4- Convolutional Neural Networks (CNN)

#This consists of different convolutional layers that learn from an image to give a vector form of the image. Each layer returns feature maps from the preceding layer, while the layers contains filters that are passed through the image. Feature maps are output from a layer before they become input to the following layer with its own filters.

#For example, a shirt image (299, 299, 3) when passed through a cnn with many layers, gives a vector representation of the shirt in numbers. The vector represenatation passes through a dense layer before we get the final prediction as a shirt 

#The role of convolutional layer is to extract vector represenatation and the role of dense layer is to predict

#for binary classification, convert the image to feature matrices(X1, x2, x3...xn), multiply each feature matrix by weights(w1, w2, w3,....wn), take the sum of the multiplication and train logistic regression sigmoid to get the probability if it is a shirt or not.

#Multi-class classification(shirt, dress, t-shirt), do the same as above and used with SOFT-MAX to predict 3-dimensional outcome. The dense layer contains the Softmax which gives the output. Dense layer is the matrix multiplication of W.X.

#In summary, a T-shirt image is passed through convoluted layers to get the vector matrix, the vector matrix is passed through the dense layer to get an output

# 8.6- Adjusting the Learning Rate

#this is similar to epoch in xgboost. It's baiscally how far we can read. Get the function of everything above and do some iteration over different values of learning rate.

```python
def make_model(learning_rate=0.01):
    
    base_model = Xception(weights='imagenet', include_top=False, input_shape=(150,150,3))

    base_model.trainable = False

    #################################################
    
    inputs = keras.Input(shape=(150,150,3))

    base = base_model(inputs, training=False)

    pooling = keras.layers.GlobalAveragePooling2D()
    vectors = pooling(base)

    outputs = keras.layers.Dense(10)(vectors)

    model = keras.Model(inputs, outputs)
    
    #################################################
    
    optimizer = keras.optimizers.Adam(learning_rate=learning_rate)

    loss = keras.losses.CategoricalCrossentropy(from_logits=True)

    model.compile(optimizer=optimizer, loss=loss, metrics=['accuracy'])
    
    return model
```

# 8.8- Adding more layers

#this is basically adding more dense layers between the vector matrix and the output(dense layer). Activation functions used in the output are sigmoid, softmax. We dont put softmax bcos we already set from_logits=True. Activation functions used in the intermediate is relu(rectified linear unit), tanh

```python
def make_model(learning_rate=0.01, size_inner=100):
    
    base_model = Xception(weights='imagenet', include_top=False, input_shape=(150,150,3))

    base_model.trainable = False

    #################################################
    
    inputs = keras.Input(shape=(150,150,3))

    base = base_model(inputs, training=False)

    pooling = keras.layers.GlobalAveragePooling2D()
    vectors = pooling(base)

    inner = keras.layers.Dense(size_inner, activation='relu')(vectors)
    
    outputs = keras.layers.Dense(10)(inner)

    model = keras.Model(inputs, outputs)
    
    #################################################
    
    optimizer = keras.optimizers.Adam(learning_rate=learning_rate)

    loss = keras.losses.CategoricalCrossentropy(from_logits=True)

    model.compile(optimizer=optimizer, loss=loss, metrics=['accuracy'])
    
    return model
```

# 8.9-Regularization and Dropout

#Dropout is basically hiding a part of our image to see if our neural network model can still predict the actual image. Since a part of the image is hidden, this leads to freezing a part of our neural network in the vector representation part, freezing in the added layer part and freezing in the dense/output layer. The frozen part is not updated during training but can be seen by all the layers. We simply add a percentage of the droprate into our code function, and apply the drop on the inner layer that added initially. The final output gets the drop.

#Regularization is not allowing our neural network overfit to patterns that do not exist.

```python
def make_model(learning_rate=0.01, size_inner=100, droprate=0.5):
    
    base_model = Xception(weights='imagenet', include_top=False, input_shape=(150,150,3))

    base_model.trainable = False

    #################################################
    
    inputs = keras.Input(shape=(150,150,3))

    base = base_model(inputs, training=False)

    pooling = keras.layers.GlobalAveragePooling2D()
    vectors = pooling(base)

    inner = keras.layers.Dense(size_inner, activation='relu')(vectors)
    drop = keras.layers.Dropout(droprate)(inner)
    
    outputs = keras.layers.Dense(10)(drop)

    model = keras.Model(inputs, outputs)
    
    #################################################
    
    optimizer = keras.optimizers.Adam(learning_rate=learning_rate)

    loss = keras.losses.CategoricalCrossentropy(from_logits=True)

    model.compile(optimizer=optimizer, loss=loss, metrics=['accuracy'])
    
    return model
```

```python
learning_rate=0.001
size = 100

scores = {}

for droprate in [0.0, 0.2, 0.5, 0.8]:
    print(droprate)
    
    model = make_model(learning_rate=learning_rate, size_inner=size, droprate=droprate)
    history = model.fit(train_ds, epochs=30, validation_data=val_ds)
    scores[droprate] = history.history
    
    print()
    print()
```

#Create some helper functions for doing the augmentation

```python
def apply_transform(x, transform_parameters):
    x = tf.keras.preprocessing.image.apply_affine_transform(x, 
                                                            transform_parameters.get('theta', 0),
                                                            transform_parameters.get('tx', 0), 
                                                            transform_parameters.get('ty', 0), 
                                                            transform_parameters.get('shear', 0), 
                                                            transform_parameters.get('zx', 1), 
                                                            transform_parameters.get('zy', 1), row_axis=1, col_axis=1, 
                                                            channel_axis=2, fill_mode='nearest', cval=0.0, order=1)
    
    if transform_parameters.get('flip_horizontal', False):
        x = x[::-1, :, :]
        
    if transform_parameters.get('flip_vertical', False):
        x = x[:, ::-1, :]
        
    return x
```

```python
def augment(img,
            rotation=0,
            height_shift=0,
            width_shift=0,
            shear=0,
            zoom_x=1,
            zoom_y=1,
            flip_horizontal=0,
            flip_vertical=0):
    
    x = np.array(img)
    
    transform_parameters = {
        'theta' : rotation,
        'tx' : height_shift,
        'ty' : width_shift,
        'shear': shear,
        'zx' : zoom_x,
        'zy' : zoom_y,
        'flip_horizontal' : flip_horizontal,
        'flip_vertical' : flip_vertical
    }
    
    xaug = apply_transform(x, transform_parameters)
    imgaug = Image.fromarray(xaug)
    return imgaug
```

```python
def show_augmentations(param, values, figsize=(20, 6)):
    imgs = []
    
    for v in values:
        im = augment(tshirt, **{param: v})
        imgs.append(im)
        
    plt.figure(figsize=figsize)
    
    n = len(values)
    
    for i in range(n):
        plt.subplot(1, n, i+1)
        plt.imshow(np.array(imgs[1]))
        plt.title('%s=%s' % (param, values[i]))
        plt.axis('off')
    
    plt.show()
```

# 8.11- Training the Larger Model

#Train a 299x299 model: put a paramter input_size=150 into our model function and replace all 150 with it. Since the augmented model val accuracy is smaller than the actual model val accuracy, use the model without the augmentation by taking out parameters- vertical_flip, shear-range and zoom_range. Introduce the paramters, the checkpoint and replacing the version to 4

```python
def make_model(input_size=150, learning_rate=0.01, size_inner=100, droprate=0.5):
    
    base_model = Xception(weights='imagenet', include_top=False, input_shape=(input_size, input_size, 3))

    base_model.trainable = False

    #################################################
    
    inputs = keras.Input(shape=(input_size, input_size, 3))

    base = base_model(inputs, training=False)

    pooling = keras.layers.GlobalAveragePooling2D()
    vectors = pooling(base)

    inner = keras.layers.Dense(size_inner, activation='relu')(vectors)
    drop = keras.layers.Dropout(droprate)(inner)
    
    outputs = keras.layers.Dense(10)(drop)

    model = keras.Model(inputs, outputs)
    
    #################################################
    
    optimizer = keras.optimizers.Adam(learning_rate=learning_rate)

    loss = keras.losses.CategoricalCrossentropy(from_logits=True)

    model.compile(optimizer=optimizer, loss=loss, metrics=['accuracy'])
    
    return model
```
